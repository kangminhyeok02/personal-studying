```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=6)#선 두꼐 
```

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='None')
```

```python
plt.plot(x,y,marker='v')
```

```python
plt.plot(x,y,marker='v',markersize=7)
```

```python
plt.plot(x,y,marker='X')
```

```python
plt.plot(x,y,marker='X',markersize=8,markeredgecolor='red')
```

```python
plt.plot(x,y,marker='X',markersize=8,markeredgecolor='red',markerfacecolor='yellow')
```

선 스타일 

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')
```

색깔

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='g')
```

다같이 포맷 

```python
plt.plot(x,y,'ro--') # color marker line style 
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go')
```

축약어

```python
plt.plot(x,y,marker='o',mfc='red',mec='blue',ms=10,ls=':') 
```

투명도

```python
plt.plot(x,y,marker='o',alpha=0.3) #투명도 (0-1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=200) # dpi 해상도 적용 
plt.plot(x,y)
```

배경색 

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```

```python
plt.figure(facecolor='y')
plt.plot(x,y)
```
