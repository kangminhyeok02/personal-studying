```python
import matplotlib.pyplot as plt 
```

1. 그래프 기본 

```python
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.title('Line Graph')
```
