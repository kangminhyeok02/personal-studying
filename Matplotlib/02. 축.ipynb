```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.title("꺽은선 그래프",fontdict={'family':'HYGungSo-Bold','size':20}) #개별 폰트 설정 가능 
```

```python
plt.xlabel('x축')
plt.ylabel('Y축')
```

```python
plt.xlabel('x축',color='red')
plt.ylabel('Y축',color='green')
```

```python
plt.xlabel('x축',loc='right') # left center right
plt.ylabel('Y축',loc='top')       # top center bottom
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
```
