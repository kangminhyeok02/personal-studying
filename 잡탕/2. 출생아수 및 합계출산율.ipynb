```python
df=df.T  # row와 column 바꿈 
```

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']= False
```

```python
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index,df['합계 출산율'])
```

```python
fig,ax1=plt.subplots(figsize=(8,3))
ax1.plot(df.index,df['출생아 수'],color='red')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis

ax2.plot(df.index,df['합계 출산율'],color='yellow')
```

```python
fig,ax1=plt.subplots(figsize=(8,3))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='red')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='yellow')
```

```python
fig,ax1=plt.subplots(figsize=(8,3))
fig.suptitle('출생아 수 및 합계출산율')
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='red')

for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+5,val,ha='center')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='yellow')
```

```python
fig,ax1=plt.subplots(figsize=(8,3))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='red')

for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+5,val,ha='center')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='yellow',marker='o',ms=10,lw=4,markeredgecolor='w',mew=2)

for idx,value in enumerate(df['합계 출산율']):
    ax2.text(idx,value+0.06,value,ha='center')
```
