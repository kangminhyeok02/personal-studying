```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#1
data_bank=pd.read_csv("C:\\Users\\pc\\Desktop\\정보\\UniversalBank.csv")
data_bank
```

```python
#1-1
Ploan = pd.value_counts(data_bank['Personal Loan'].values, sort=True)
Ploan.plot.bar()
Ploan.plot.bar(grid=True, figsize=(10,8), fontsize=15)
plt.xlabel('Personal Loan')
plt.ylabel('Number of customers')
plt.show()
```

1-1번에서 그린 그래프는 Personal Loan의 종류에 따른 고객 수를 나타낸 barplot이다. 카테고리별 데이터의 개수를 나타내는 데 제일 적합한 plot이라 생각해서 barplot으로 나타냈다. 개인 대출을 신청한 기록이 있는 고객과 그렇지 않은 고객의 비율이 대략 1:9인 것을 그래프를 통해 알 수 있다.

```python
#1-2
plt.figure(figsize=(12,12))
plt.subplot(6,2,1)
sns.kdeplot(data = data_bank,x = 'Age',hue = 'Personal Loan').set_title("Age")
plt.subplot(6,2,2)
sns.kdeplot(data = data_bank, x = 'Experience',hue = 'Personal Loan').set_title("Experience")
plt.subplot(6,2,3)
sns.kdeplot(data = data_bank, x = 'Income', hue = 'Personal Loan').set_title("Income")
plt.subplot(6,2,4)
sns.kdeplot(data = data_bank, x = 'Family', hue = 'Personal Loan').set_title("Family")
plt.subplot(6,2,5)
sns.kdeplot(data = data_bank, x = 'CCAvg', hue = 'Personal Loan').set_title("CCAvg")
plt.subplot(6,2,6)
sns.kdeplot(data = data_bank, x = 'Mortgage', hue = 'Personal Loan').set_title("Mortgage")

plt.tight_layout()
plt.show()
```

1-2의 그래프는 개인 대출의 여부에 따른 나이, 경력, 수입, 가구, 평균 한 달 신용카드 지출, 저당 등의 여섯 가지 요인들의 분포도를 나타낸다. 대략 나이로는 20세 중반 ~ 70세, 전문 경력은 5 ~ 45년 구간에서 밀도가 높다고 나오는데, 결국 연령대나 사회 경험의 기간과는 관계없이 개인 대출이 이루어진다 볼 수 있다.
수입의 경우, 개인 대출 경험자들은 100 초중반과 200 사이에, 무경험자들은 수입이 주로 100 이하에 머무는데, 대출을 한 사람들 중에서는 고수익자들이 상대적으로 많다고 볼 수 있다. 한편, 가구 수에 따른 개인 대출자들의 빈도를 따져보면, 그래프를 그리기 전 나의 예상대로 가구 수가 많아질수록 대출자들이 많아진다. 아무래도 부양 가구가 많아지는 3, 4인 가구의 지표가 1, 2인 가구의 지표보다 상대적으로 크다. CCAvg, 즉 평균 한 달 신용카드 지출 그래프는 수입 그래프와 모양이 비교적 비슷한데, 대체적으로 카드 지출이 커질수록 대출을 고려하는 빈도가 커지는 경향을 보인다. 

```python
#2
data_set=pd.read_csv("C:\\Users\\pc\\Desktop\\정보\\kbobattingdata.csv")
data_set
```

```python
#scatter plot

x=data_set[['year']]
y=data_set[['homeruns']]

plt.scatter(x,y)
plt.xlabel('year')
plt.ylabel('homeruns')
plt.title('Homeruns VS Year')
plt.show()
```

x축을 년도 y축을 홈런 개수로 하는 산점도 입니다. 산점도를 통해 두 변수 년도와 홈런 개수 사이의 관계에 대해 알 수 있습니다. 년도가 증가할수록 홈런의 개수가 증가한다는 것을 알 수 있습니다.

```python
#bar plot
lg_data=data_set[data_set['team']=='LG Twins']
x=list(lg_data['year'][0:5])
y=list(lg_data['homeruns'][0:5])
plt.figure(figsize = (7,4))
plt.bar(x, y, alpha = 0.7, color = 'red')
plt.title('LG Twins homeruns')
plt.xlabel('year')
plt.ylabel('Homeruns')
plt.show()
```

가장 좋아하는 팀인 lg와 야구를 즐겨 보기 시작하던 2017년부터의 LG의 홈런 개수를 bar plot을 이용해 그려보았습니다. 각 바의 색은 LG를 상징하는 빨간색으로 표현하였습니다. 그래프 상에서 전체적으로 보면 2019년 홈런 갯수가 적은걸 알 수 있습니다.

```python
#Line plot

x=lg_data[['year']]
y_1=lg_data[['hits']]
y_2=lg_data[['strikeouts']]

plt.plot(x, y_1,label='hits', color='blue', alpha = 0.3) 
plt.plot(x, y_2, label='strikeouts', color='red', alpha = 0.7)
plt.legend()
plt.xlabel('year')
plt.ylabel('hits')
plt.title('HitS VS Year')
plt.show()
```

LG 데이터를 기준으로 안타와 삼진의 관계에 대하여 직선 그래프를 그려 보았다. 실제로 안타갯수가 많을때는 삼진 갯수도 많고 적을 떄는 삼진 갯수도 적다는 걸 알 수 있다. 이는 적극적인 타격은 불가피하게 많은 삼진으로 이어진다는 추론이 가능하였습니다. 

```python
#Histogram
compare_data=data_set[(data_set['team']=='LG Twins') | (data_set['team']=='Hanwha Eagles')]
compare_data
sns.histplot(x='homeruns',data=compare_data,hue='team',element='poly')
```

LG Twins와 약팀이 환화 이글스의 홈런 갯수를 히스토그램으로 그려 비교 해보았다. 예상과 같이 환화는 홈런 갯수가 적은 약팀이었다

```python
#count plot 
data_2014=data_set[data_set['year']>=2014]
sns.countplot(x='games',data=data_2014,hue='team')
```

count plot 의 사용해서 2014년 이후 팀별로 경기 수를 표현해보았다. 그래프에서 보면 알 수 있듯이 2014년 끝나고 2015년부터 새로운 팀인 kt wiz의 합류로 팀별 경기수가 144경기로 늘어났다는 사실의 알 수 있다. 그래프 상에서 143경기인 case가 1번 있는데 이는 2021년의 값으로 데이터 작성시점에 아직 1경기가 남은 경우여서 다음과 같이 표현되었다. 

```python
#Box plot
sns.boxplot(data=compare_data, x='team',y='homeruns')
plt.title('LG Twins vs Hanwha Eagles')
plt.xlabel('team')
plt.ylabel('Homeruns')
```

bar plot의 이용해 LG와 환화의 홈런 갯수를 그려 보았다. 최대,최소,평균 을 보기에 아주 유리한 형태이다. 또 홈런과 같은 연속적인 변수를 표현하는 것에도 저갑한 그래프이다. 

```python
#Heat map 
data_heat=data_set[['hits','homeruns','strikeouts']]
corr_df = data_heat.corr(method='pearson')
plt.matshow(corr_df)
sns.heatmap(corr_df, annot=True)
plt.show()
```

홈런,삼진,안타의 갯수 사이의 상관관계를 보여주는 heat map의 그려보았다. 홈런과 삼진의 상관게수의 절대값은 0.62 로 어느정도 꽤 높은 상관관계가 있다는 것을 알 수 있습니다. 삼진과 안타의 상관계수의 절대값은 0.78로 매우 높은 상관관계를 가집니다. 이는 상식적으로 적극적으로 타격에 임하는 타자의 삼진갯수는 어쩔수 없이 높을 것이라는 상식에 부합하는 결과입니다. 

```python
#lmplot
data_2015=data_set[data_set['year']>=2015]
data_find=data_2015[(data_2015['team']=='LG Twins') | (data_2015['team']=='Hanwha Eagles') | (data_2015['team']=='Lotte Giants') | (data_2015['team']=='Doosan Bears')]
sns.lmplot(x='year',y='homeruns',data=data_find,col='team',col_wrap=2)
```

2015년 이후 두산,롯데,LG,환화 4개의 팀에 대한 년도별 홈런수 변화를 lmplot의 이요해 그려 보았다. LG 를 제외하고 다른 3개의 팀은 홈런의 갯수가 감소하는 추세를 보이지만 유일하게 LG만 갯수가 유지되는 모양이다. 사실 최근 야구계에서 투수의 실력 향상으로 인해 점수가 많이 나오지 않는 지루한 경기의 문제점이 지적되고 있는데 실제로 홈런갯수가 감소하는 위 자료가 이를 어느정도 보여준다고 생각합니다.
